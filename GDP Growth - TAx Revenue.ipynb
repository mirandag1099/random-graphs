```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/miran/Desktop/finanzas.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['tax_revenue'] = df['tax_revenue'] * 100
```

```python
df.head()
```

```python
fig, ax = plt.subplots()
ax.plot(df['year'], df['gdp_growth'], color='blue')
ax.set_xlabel('Año')
ax.set_ylabel('% crecimiento PIB', color='blue')
ax2 = ax.twinx()
ax2.plot(df['year'], df['tax_revenue'], color='red')
ax2.set_ylabel('Variación porcentual ingresos por T', color='red')
plt.show()

fig.savefig('prueba', format='jpeg', dpi=1500)
```

```python
plt.rcParams.update({'font.family': 'Book Antiqua'})
plt.figure(figsize=(8,6))

fig,ax=plt.subplots()
ax.plot(df['year'], df['gdp_growth'], label="variación % PIB")
ax.set_title(label="Evolución del PIB y los ingresos por impuestos (2011-2020)")
ax.set_xlabel('Año')
ax.set_ylabel('%')
ax.plot(df['year'], df['tax_revenue'])
plt.show()
```

```python
#Grafico Final
plt.figure(figsize=(8,6))

plt.plot(df['year'], df['gdp_growth'], color='blue', label='PIB')
plt.plot(df['year'], df['tax_revenue'], color='orange', label='Impuestos')
plt.xlabel('Año', fontsize=13)
plt.ylabel('%', fontsize=13)
plt.suptitle("PIB e Ingresos por impuestos en Nicaragua (2011-2020)", fontsize=15)
plt.legend(loc='upper left')
plt.ylim(-6, 12.5)

txt = "Nota: Elaboración propia con datos del Banco Mundial"
plt.figtext(0.5, 0.01, txt, wrap=True, horizontalalignment='center', fontsize=9)

plt.savefig('finanzas.png', dpi=2000)
```
